```python
def calculate_P(): #Wahrscheinlichkeit, dass das Fahrrad funktioniert
    P = [0] * 100  #Ein Semester, 100 Tage
    P[0] = 1  # Am ersten Tag des Semesters funktioniert das Fahrrad.

    for i in range(1, 100):
        P[i] = 0.95 * P[i-1] + 0.6 * (1 - P[i-1])

    return P

result = calculate_P()
print(result[99])  #Wahrscheinlichkeit, dass das Fahrrad funktioniert am 100.Tag
```

```python
import matplotlib.pyplot as plt

def calculate_P():
    P = [0] * 100
    P[0] = 1

    for i in range(1, 100):
        P[i] = 0.95 * P[i-1] + 0.6 * (1 - P[i-1])

    return P

result = calculate_P()

plt.plot(range(100), result)
plt.xlabel('i (Tage-1)')
plt.ylabel('P(i)')
plt.title('Wahrscheinlichkeit, dass das Fahrrad funktioniert')
plt.show()
```
